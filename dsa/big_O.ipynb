```python
from time import time
import matplotlib.pyplot as plt
```

```python
def time_complexity(f, n : None = 5):

    time_taken = [0] * n
    sizes = list(range(1,n+1))
    for i in range(n) : 

        arr_n = [j+1 for j in range(i+1)]
        start_time = time()
        f(arr_n)
        end_time = time()
        time_taken[i] = end_time - start_time

    plt.figure(figsize=(10,10))
    plt.plot(sizes, time_taken, 'b-o')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Time Taken (s)')
    plt.grid(True)
    plt.show
    #eturn time_taken
        
```

```python
arr = [2,4,6,8,10]

def prefix_average_1(S : list) : 

    n = len(S) #5   
    A = [0] * n

    for i in range(n): # i = 0,1,2,3,4
        
        total = 0 

        for j in range(i+1) : # if i = 3, j = 0,1,2,3
            # j = 0, total = 0 + 2 = 2
            # j = 1, total = 2 + 4 = 6


            total = total + S[j]

        
        A[i] = total / (i + 1)

    return A

def prefix_average_2(S : list): 
    
    n = len(S)
    A = [0] * n

    total = 0 

    for i in range(n) : 
        total = total + S[i]
        A[i] = total / (i+1)

    return A


print(prefix_average_1(arr))
print(prefix_average_2(arr))

```

```python
time_complexity(prefix_average_1, 1000)
```

```python
time_complexity(prefix_average_2,1000)
```

```python
# element uniqueness 

def if_unique(L : list) : 

    n = len(L)
    for i in range(n): 
        
        for j in range(i+1,n):  
                if L[i] == L[j]:
                    return False
    return True


arr1 = [1,2,3,5,5]
if_unique(arr1)

time_complexity(if_unique,1000)
```

```python
### using python sorted

def if_unique_2(L : list) : 

    SL = list(sorted(L))

    for i in range(len(SL)-1): 
        if SL[i] == SL[i+1] : 
            return False 
    return True


if_unique_2([1,2,3,4,5,5])

time_complexity(if_unique_2,1000)
```

A sequence S contains n−1 unique integers in the range [0,n−1], that
 is, there is one number from this range that is not in S. Design an O(n)
time algorithm for finding that number. You are only allowed to use O(1)
 additional space besides the sequence S itself.

```python
def find_missing_number(S : list) : 
    n = len(S)

    for i in range(0,n+1):
       if i not in S:
           return i
    
S = [0,1,2,3]
find_missing_number(S)
        
```

```python
warmtones = ['red','orange','yellow']
palette = warmtones
palette.append("blue")
print(warmtones)
```

```python
references = []
for i in range(len(palette)):
    references.append((id(palette[i]), id(warmtones[i])))
print(references)

print(id(warmtones) == id(palette))
```

```python
n = 1000000
data = []
iteration = [i for i in range(n)]
time_taken = []
array_memory_space = []
for i in range(n):
    start_time = time()
    data.append(None)
    end_time = time()
    time_taken.append(end_time-start_time)
    array_memory_space.append(getsizeof(data))


plt.figure(figsize=(10,10))
plt.plot(iteration, array_memory_space, 'b-o')
plt.xlabel('iteration (n)')
plt.ylabel('array memory space (bytes)')
plt.grid(True)
plt.show
```
